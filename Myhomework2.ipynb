```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

# Fetch S&P 500 data from Yahoo Finance for 10 years
df = yf.download("^GSPC", start="2013-08-01", end="2023-08-01")

# Calculate daily returns and annualized volatility
df['Returns'] = df['Close'].pct_change()
df['Volatility'] = df['Returns'].rolling(window=252).std() * np.sqrt(252)

# Plot the S&P 500 prices
plt.figure(figsize=(14, 7))
plt.subplot(2, 1, 1)
plt.plot(df.index, df['Close'], label='S&P 500 Prices', color='blue')
plt.title('S&P 500 Prices')
plt.ylabel('Price')
plt.legend()

# Plot the annualized volatility
plt.subplot(2, 1, 2)
plt.plot(df.index, df['Volatility'], label='Annualized Volatility', color='red')
plt.title('S&P 500 Annualized Volatility')
plt.ylabel('Annualized Volatility')
plt.xlabel('Year')
plt.legend()

# Plot returns
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Returns'], label='Daily Returns')
plt.title('S&P 500 Stock Returns')
plt.ylabel('Return')
plt.xlabel('Date')
plt.legend()
plt.show()

plt.tight_layout()
plt.show()
```
